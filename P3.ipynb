```python
# Número de linhas antes da limpeza
linhas_antes = len(df)

# Remove somente linhas totalmente vazias
df = df.dropna(how="all").reset_index(drop=True)

# Número de linhas depois da limpeza
linhas_depois = len(df)

# Quantas foram removidas
removidas = linhas_antes - linhas_depois

print(f"Linhas antes: {linhas_antes}")
print(f"Linhas depois: {linhas_depois}")
print(f"Total removidas: {removidas}")
```

```python
import numpy as np

# Cria uma cópia temporária para checar
df_temp = df.replace(r'^\s*$', np.nan, regex=True)

# Localiza linhas totalmente vazias (todas as colunas == NaN)
linhas_ocultas = df_temp[df_temp.isna().all(axis=1)]

# Conta quantas são
qtd_ocultas = len(linhas_ocultas)

print(f"Linhas totalmente vazias (incluindo as 'ocultas'): {qtd_ocultas}")

# Se quiser ver os índices originais dessas linhas
print("Índices das linhas vazias:", linhas_ocultas.index.tolist())
```

```python
import numpy as np

# 1. Substituir células só com espaços por NaN
df = df.replace(r'^\s*$', np.nan, regex=True)

# 2. Contar quantas linhas ficaram totalmente vazias
linhas_ocultas = df[df.isna().all(axis=1)]
qtd_ocultas = len(linhas_ocultas)

print(f"Linhas totalmente vazias (incluindo ocultas): {qtd_ocultas}")
print("Índices removidos:", linhas_ocultas.index.tolist())

# 3. Excluir essas linhas e resetar índice
df = df.dropna(how="all").reset_index(drop=True)

print(f"Novo tamanho da base: {len(df)} linhas")
```
